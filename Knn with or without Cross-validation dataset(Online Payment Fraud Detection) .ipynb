## Knn with or without Cross-validation

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\sujat\Downloads\Online Payment fraud detection.csv")
data
```

```python
data1=data.iloc[:,[1,2,4,5,7,8,9]]
data1
```

```python
from sklearn.preprocessing import LabelEncoder
l1=LabelEncoder()
data1["type"]=l1.fit_transform(data1["type"])
```

```python
data1
```

```python
data1["isFraud"].value_counts()
```

```python
x=data1.iloc[:,:-1].values
y=data1.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
s1=StandardScaler()
s2=s1.fit_transform(x)
```

```python
data2=pd.DataFrame(s2)
data2
```

```python
x1=data2.iloc[:,:-1].values
y1=y
```

```python
from imblearn.over_sampling import RandomOverSampler
r1=RandomOverSampler()
x_data,y_data=r1.fit_resample(x1,y1)
```

```python
from collections import Counter
print(Counter(y_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k1=KNeighborsClassifier()
k1.fit(x_train,y_train)
```

```python
y_pred=k1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_pred,y_test)*100
print(ac)
```

## With Cross-Validation

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,random_state=11,shuffle=True)
kf.get_n_splits(x1)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
score=cross_val_score(k1,x_train,y_train,cv=kf)*100
pred=cross_val_predict(k1,x_test,y_test)
print(score)
print(np.mean(score))
print("knn:",pred)
```
